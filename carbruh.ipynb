```python
#Functiopn to remove outlier 
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
petrol_clean = remove_outlier(petrol,'co2_g_per_km')
petrol_clean.head()
```

```python
X = df3.drop(columns=['fuel_type'])
```

```python
y = df3['fuel_type'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
y[0:5]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)

# Fit the classifier to the data
knn.fit(X_train,y_train)
```

```python
#slice of the prediction
knn.predict(X_test)[0:5]
```

```python
#Accuraacy of current run
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
#create a new KNN model
knn_cv = KNeighborsClassifier(n_neighbors=3)

#train model with cv of 5 
cv_scores = cross_val_score(knn_cv, X, y, cv=5)

#print each cv score (accuracy) and average them
print(cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))
```

```python
from sklearn.model_selection import GridSearchCV

#create new a knn model
knn2 = KNeighborsClassifier()

#create a dictionary of all values we want to test for n_neighbors
param_grid = {'n_neighbors': np.arange(1, 25)}

#use gridsearch to test all values for n_neighbors
knn_gscv = GridSearchCV(knn2, param_grid, cv=5)

#fit model to data
knn_gscv.fit(X, y)
```

HIDDEN INFERENCES

A.We can observe that the value of particulates for diesel vehicles is very low compared to that of petrol vehicles    
  even when diesel is a denser fuel with a higher calorific value this has to do with two factors.
  
  1.The way with which diesel is combusted in the engine - Diesel isn'nt ignited by spark plugs but rather by                      
  compression of inlet air-fuel mixture which results in a cleaner combustion of the fuel.
  
  2.Diesel engine for a given capacity is more efficient than their petrol counter-parts as all diesel engines by
  design are turbocharged.
